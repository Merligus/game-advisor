```python
import pandas as pd
```

```python
game_review_df = pd.read_csv("../data/multi_decade_video_game_review_dataset.csv")
relevant_columns = [
    "Review Title",
    "Gaming Console name",
    "Time Since Review",
    "Rating/10",
]
# filter na values
game_review_df = game_review_df.dropna(subset=["Rating/10"])
# filter columns
game_review_df = game_review_df[relevant_columns]
game_review_df
```

```python
hltb_dataset_df = pd.read_csv("../data/hltb_dataset.csv")
# get only games
hltb_dataset_df = hltb_dataset_df[hltb_dataset_df["type"] == "game"]
relevant_columns = [
    "name",
    "platform",
    "genres",
    "developer",
    "publisher",
    "release_year",
    "main_story",
    "main_plus_sides",
    "completionist",
]
# drop lines with na values
hltb_dataset_df = hltb_dataset_df.dropna(
    subset=["main_story", "main_plus_sides", "completionist"]
)
# filter columns
hltb_dataset_df = hltb_dataset_df[relevant_columns]
hltb_dataset_df
```

```python
metacritics_games_df = pd.read_csv("../data/metacritics_games.csv")
relevant_columns = [
    "title",
    "releaseDate",
    "genres",
    "description",
    "platforms",
    "metascore",
    "userscore",
    "developer",
    "publisher",
]
# drop lines with na values
metacritics_games_df = metacritics_games_df.dropna(subset=["metascore", "userscore"])
# filter columns
metacritics_games_df = metacritics_games_df[relevant_columns]
metacritics_games_df
```

```python
rawg_games_dataset_df = pd.read_csv("../data/rawg_games_dataset.csv")
relevant_columns = [
    "name",
    "released",
    "rating",
    "platforms",
    "developers",
    "genres",
    "tags",
    "publishers",
    "description",
    "description_raw",
    "metacritic",
]
# filter columns
rawg_games_dataset_df = rawg_games_dataset_df[relevant_columns]
rawg_games_dataset_df
```
